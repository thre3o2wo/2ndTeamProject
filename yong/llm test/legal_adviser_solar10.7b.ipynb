```python
import os
import cohere
from dotenv import load_dotenv
from langchain_upstage import UpstageEmbeddings
from langchain_pinecone import PineconeVectorStore
from pinecone import Pinecone
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 1. í™˜ê²½ ë³€ìˆ˜ ë¡œë“œ
load_dotenv(override=True)

PINECONE_API_KEY = os.getenv("PINECONE_API_KEY")
UPSTAGE_API_KEY = os.getenv("UPSTAGE_API_KEY")
COHERE_API_KEY = os.getenv("COHERE_API_KEY")

# =========================================================
# 2. [ì „ì²˜ë¦¬] í‚¤ì›Œë“œ ì •ê·œí™” (Inline Injection ë°©ì‹)
# =========================================================
def normalize_query(user_query):
    """
    ì‚¬ìš©ì ì¿¼ë¦¬ ë‚´ì˜ ì¼ìƒ ìš©ì–´ë¥¼ ë²•ë¥  ìš©ì–´ë¡œ ë³€í™˜í•˜ì—¬,
    ë¬¸ë§¥ ë‚´ì— 'ì¼ìƒì–´(ë²•ë¥ ì–´)' í˜•íƒœë¡œ ì£¼ì…í•©ë‹ˆë‹¤.
    ì˜ˆ: "ì§‘ì£¼ì¸ì´ ì›”ì„¸ë¥¼ ì˜¬ë ¤" -> "ì§‘ì£¼ì¸(ì„ëŒ€ì¸)ì´ ì›”ì„¸(ì°¨ì„)ë¥¼ ì˜¬ë ¤(ì¦ì•¡)"
    """
    
    # 1. ë§ˆìŠ¤í„° í‚¤ì›Œë“œ ë”•ì…”ë„ˆë¦¬ (ê¸´ ë‹¨ì–´ë¶€í„° ë§¤ì¹­ë˜ë„ë¡ ì •ë ¬ í•„ìš”)
    # Key: ì¼ìƒì–´, Value: ë²•ë¥ ì–´
    keyword_dict = {
        # [ì£¼ì²´]
        "ì§‘ì£¼ì¸": "ì„ëŒ€ì¸", "ì£¼ì¸": "ì„ëŒ€ì¸", "ê±´ë¬¼ì£¼": "ì„ëŒ€ì¸", "ì„ëŒ€ì—…ì": "ì„ëŒ€ì¸",
        "ì„¸ì…ì": "ì„ì°¨ì¸", "ì›”ì„¸ì…ì": "ì„ì°¨ì¸",
        "ë¶€ë™ì‚°": "ê³µì¸ì¤‘ê°œì‚¬", "ì¤‘ê°œì¸": "ê³µì¸ì¤‘ê°œì‚¬",
        
        # [ê¸ˆì „]
        "ë³´ì¦ê¸ˆ": "ì„ì°¨ë³´ì¦ê¸ˆ", "ì „ì„¸ê¸ˆ": "ì„ì°¨ë³´ì¦ê¸ˆ", 
        "ì›”ì„¸": "ì°¨ì„", "ë°©ì„¸": "ì°¨ì„",
        "ê´€ë¦¬ë¹„": "ê´€ë¦¬ë¹„ìš©",
        
        # [í–‰ìœ„ - ëˆ]
        "ì˜¬ë ¤": "ì¦ì•¡", "ì¸ìƒ": "ì¦ì•¡", "ë” ë‹¬ë¼ê³ ": "ì¦ì•¡", "ë” ë‚´ë¼ê³ ": "ì¦ì•¡",
        "ê¹ì•„": "ê°ì•¡", "ë‚´ë ¤": "ê°ì•¡", "í• ì¸": "ê°ì•¡",
        "ëŒë ¤ì£¼ì§€": "ë°˜í™˜", "ì•ˆ ì¤˜": "ë°˜í™˜ê±°ë¶€", "ë–¼ë¨¹": "ë¯¸ë°˜í™˜",
        
        # [í–‰ìœ„ - ê±°ì£¼/ê³„ì•½]
        "ê³„ì•½ì„œ": "ì„ëŒ€ì°¨ê³„ì•½ì¦ì„œ", "íŠ¹ì•½": "íŠ¹ì•½ì‚¬í•­",
        "ì´ì‚¬": "ì£¼íƒì˜ì¸ë„", "ì „ì…": "ì£¼ë¯¼ë“±ë¡/ëŒ€í•­ë ¥",
        "ì‚´ê³ ": "ì ìœ ", "ê±°ì£¼": "ì ìœ ",
        "ë‚˜ê°€ë¼": "ê³„ì•½ê°±ì‹ ê±°ì ˆ/ëª…ë„", "ë‚˜ê°€ë¼ê³ ": "ê³„ì•½ê°±ì‹ ê±°ì ˆ/ëª…ë„", "ë¹„ì›Œ": "ëª…ë„", "ì«“ê²¨": "ëª…ë„/ëŒ€í•­ë ¥",
        "ë°© ë¹¼": "ê³„ì•½í•´ì§€", "ë‚˜ê°ˆê²Œ": "ê³„ì•½í•´ì§€", "í•´ì§€": "ê³„ì•½í•´ì§€",
        "ì—°ì¥": "ê³„ì•½ê°±ì‹ ", "ë” ì‚´": "ê³„ì•½ê°±ì‹ ", "ì¬ê³„ì•½": "ê³„ì•½ê°±ì‹ ",
        "ìˆ˜ë¦¬": "ìˆ˜ì„ ì˜ë¬´/í•„ìš”ë¹„", "ê³ ì³": "ìˆ˜ì„ ì˜ë¬´", "ë¬¼ ìƒˆ": "ëˆ„ìˆ˜", "ê³°íŒ¡ì´": "í•˜ì",
        "ì²­ì†Œ": "ì›ìƒíšŒë³µì˜ë¬´/í†µìƒì˜ì†ëª¨",
        "ì²­ì†Œë¹„": "ì›ìƒíšŒë³µì˜ë¬´/ë¹„ìš©ìƒí™˜",
        "ì…ì£¼ì²­ì†Œ": "ì›ìƒíšŒë³µì˜ë¬´",
        "í‡´ì‹¤": "ì„ëŒ€ì°¨ì¢…ë£Œ/ì›ìƒíšŒë³µ",
        "ê³µì œ": "ë³´ì¦ê¸ˆë°˜í™˜ê±°ë¶€",
        "ë¬´ì¡°ê±´": "ë¶ˆê³µì •íŠ¹ì•½/ê°•í–‰ê·œì •ìœ„ë°˜",
        "ì›ìƒë³µêµ¬": "ì›ìƒíšŒë³µì˜ë¬´"
    }
    
    normalized = user_query
    
    # 2. ì¹˜í™˜ ë¡œì§ (ë‹¨ì–´ê°€ ê²¹ì¹  ê²½ìš° ê¸´ ë‹¨ì–´ ìš°ì„  ë§¤ì¹­ì„ ìœ„í•´ í‚¤ ê¸¸ì´ë¡œ ì •ë ¬)
    # ì˜ˆ: 'ì „ì„¸ê¸ˆ'ì´ 'ì „ì„¸'ë³´ë‹¤ ë¨¼ì € ì²˜ë¦¬ë˜ì–´ì•¼ í•¨
    sorted_keys = sorted(keyword_dict.keys(), key=len, reverse=True)
    
    for key in sorted_keys:
        if key in normalized:
            legal_term = keyword_dict[key]
            # ì´ë¯¸ ë³€í™˜ëœ ë¶€ë¶„(ê´„í˜¸ ì•ˆ)ì€ ê±´ë“œë¦¬ì§€ ì•Šë„ë¡ ì²´í¬í•  ìˆ˜ë„ ìˆìœ¼ë‚˜,
            # ê°„ë‹¨í•˜ê²ŒëŠ” replaceë¥¼ ì‚¬ìš©í•˜ë˜, ë¬´í•œ ë£¨í”„ ë°©ì§€ë¥¼ ìœ„í•´ í•œ ë²ˆë§Œ ì‹¤í–‰í•˜ê±°ë‚˜
            # "ì§‘ì£¼ì¸" -> "ì§‘ì£¼ì¸(ì„ëŒ€ì¸)"ìœ¼ë¡œ ë°”ê¾¸ë©´ ë‹¤ìŒ ë£¨í”„ì—ì„œ "ì£¼ì¸"ì´ ë˜ ë§¤ì¹­ë  ìˆ˜ ìˆìŒ.
            # ì´ë¥¼ ë°©ì§€í•˜ê¸° ìœ„í•´, ì—¬ê¸°ì„œëŠ” ë‹¨ìˆœ replaceë¥¼ ì“°ë˜
            # ì´ë¯¸ (ë²•ë¥ ì–´)ê°€ ë¶™ì–´ìˆëŠ”ì§€ í™•ì¸í•˜ëŠ” ë¡œì§ì€ ë³µì¡í•´ì§€ë¯€ë¡œ
            # ê°€ì¥ ì§ê´€ì ì¸ 'ë‹¨ìˆœ ì¹˜í™˜' í›„ 'ì¤‘ë³µ ì •ì œ'ëŠ” LLMì—ê²Œ ë§¡ê¹ë‹ˆë‹¤.
            
            # í•µì‹¬: "ë‹¨ì–´" -> "ë‹¨ì–´(ë²•ë¥ ìš©ì–´)" í˜•íƒœë¡œ ë³€í™˜
            # ë‹¤ë§Œ, ì´ë¯¸ ë³€í™˜ëœ ë‹¨ì–´ ë‚´ë¶€ì˜ ê¸€ìê°€ ë˜ ë§¤ì¹­ë˜ëŠ” ê²ƒì„ ë§‰ê¸° ìœ„í•´ ì„ì‹œ í† í°ì„ ì“¸ ìˆ˜ë„ ìˆìŒ.
            # ì—¬ê¸°ì„œëŠ” íŒŒíŠ¸ë„ˆë‹˜ì˜ ì´í•´ë¥¼ ë•ê¸° ìœ„í•´ ì§ê´€ì ìœ¼ë¡œ ì²˜ë¦¬í•©ë‹ˆë‹¤.
            
            # "ì£¼ì¸"ì´ "ì§‘ì£¼ì¸"ì˜ ì¼ë¶€ë¡œ ë§¤ì¹­ë˜ëŠ” ê²ƒì„ ë°©ì§€í•˜ê¸° ìœ„í•´
            # í•´ë‹¹ keyê°€ ë‹¤ë¥¸ keyì˜ ë¶€ë¶„ì§‘í•©ì´ ì•„ë‹Œ ê²½ìš°ë§Œ ì²˜ë¦¬í•˜ê±°ë‚˜,
            # replace ì‹œ ì¡°ì‹¬í•´ì•¼ í•©ë‹ˆë‹¤. 
            
            # ì•ˆì „í•œ ë°©ë²•: ë¬¸ì¥ì„ í›¼ì†í•˜ì§€ ì•ŠëŠ” ì„ ì—ì„œ ê´„í˜¸ ë³‘ê¸°
            target = f"{key}({legal_term})"
            if target not in normalized: # ì´ë¯¸ ë³€í™˜ëœ ê²Œ ì•„ë‹ˆë©´
                normalized = normalized.replace(key, target)
                
    return normalized

# =========================================================
# 3. [ê²€ìƒ‰ & Rerank] í•˜ì´ë¸Œë¦¬ë“œ ê²€ìƒ‰ (ë³€ê²½ ì—†ìŒ)
# =========================================================
def legal_hybrid_retrieval(query, pinecone_store, k_fetch=30, k_final=10, score_threshold=0.25):
    print(f"ğŸ” [1ì°¨ ê²€ìƒ‰] Pinecone íƒìƒ‰: '{query}'")
    try:
        initial_docs = pinecone_store.similarity_search(query, k=k_fetch)
    except Exception as e:
        print(f"âŒ ê²€ìƒ‰ ì‹¤íŒ¨: {e}")
        return []
    
    if not initial_docs:
        return []

    # Cohere Rerank
    docs_content = [doc.page_content for doc in initial_docs]
    co = cohere.Client(api_key=COHERE_API_KEY)
    
    try:
        rerank_results = co.rerank(
            model="rerank-multilingual-v3.0",
            query=query, # ì •ê·œí™”ëœ ì¿¼ë¦¬ ê·¸ëŒ€ë¡œ ì‚¬ìš©
            documents=docs_content,
            top_n=k_fetch
        )
    except Exception as e:
        print(f"âš ï¸ Rerank ì‹¤íŒ¨: {e}")
        return initial_docs[:k_final]

    relevant_docs = []
    for r in rerank_results.results:
        if r.relevance_score < score_threshold:
            continue
        doc = initial_docs[r.index]
        doc.metadata['relevance_score'] = r.relevance_score
        relevant_docs.append(doc)

    final_docs = sorted(
        relevant_docs, 
        key=lambda x: int(x.metadata.get('priority', 99))
    )
    
    return final_docs[:k_final]

# =========================================================
# 4. [í¬ë§·íŒ…]
# =========================================================
def format_legal_docs(docs):
    display_html = '<div class="legal-docs-container mb-3">'
    llm_context = ""
    
    for i, doc in enumerate(docs):
        src_title = doc.metadata.get('src_title', 'ë²•ë ¹')
        title = doc.metadata.get('title', '')
        
        # ì›ë³¸ì˜ ë”ë¸” ì—”í„°(\n\n)ë¥¼ ì‹±ê¸€ ì—”í„°(\n)ë¡œ ì¹˜í™˜
        content = doc.page_content.strip().replace("\n\n", "\n")
        
        priority = doc.metadata.get('priority', 99)
        
        # LLM ì…ë ¥ìš© (ê°€ë…ì„±ì„ ìœ„í•´ ì¡°í•­ ì‚¬ì´ì—ëŠ” \n\n ìœ ì§€)
        llm_context += f"ì°¸ê³ ë¬¸ì„œ {i+1}.\nì¶œì²˜: {src_title}\nì¡°í•­: {title}\në‚´ìš©: {content}\n\n"
        
        # HTMLìš©
        badge_color = "bg-danger" if priority == 1 else "bg-secondary"
        display_html += f"""
        <div class="card mb-2 shadow-sm">
            <div class="card-header d-flex justify-content-between align-items-center py-2">
                <span class="fw-bold text-dark">{title}</span>
                <span class="badge {badge_color}">{src_title}</span>
            </div>
            <div class="card-body py-2 bg-light">
                <p class="card-text text-secondary" style="white-space: pre-line; font-size: 0.9rem;">{content}</p>
            </div>
        </div>
        """
    display_html += '</div>'
    
    return {"display_html": display_html, "llm_context": llm_context}

# =========================================================
# 5. [ìƒì„±] LLM ë‹µë³€ ìƒì„± (í”„ë¡¬í”„íŠ¸ ê°•í™”)
# =========================================================
# [ìˆ˜ì •ëœ ë””ë²„ê¹…ìš© í•¨ìˆ˜] ìŠ¤íŠ¸ë¦¬ë° ì¶œë ¥ ì ìš©
def generate_legal_advice(normalized_query, context_text, stream_output=True):
    """
    Args:
        stream_output (bool): Trueë©´ ì‹¤ì‹œê°„ ì¶œë ¥(íƒ€ìê¸° íš¨ê³¼), Falseë©´ ì™„ì„± í›„ í•œ ë²ˆì— ë°˜í™˜
    """
    if stream_output:
        print(f"ğŸ¤– [LLM ì…ë ¥] Query: {normalized_query}")
        print("â³ ëª¨ë¸ ë‹µë³€ ìƒì„± ì¤‘ (Streaming ON)...")
    else:
        print("â³ ëª¨ë¸ ë‹µë³€ ìƒì„± ì¤‘ (Streaming OFF - ì ì‹œë§Œ ê¸°ë‹¤ë ¤ì£¼ì„¸ìš”)...")
    
    # [í”„ë¡¬í”„íŠ¸ ìˆ˜ì •] í•œì ì‚¬ìš© ê¸ˆì§€ ë° í˜„ëŒ€ì–´ ì‚¬ìš© ê°•ì œ
    system_prompt = """
    ë‹¹ì‹ ì€ ëŒ€í•œë¯¼êµ­ ë¶€ë™ì‚° ë²•ë¥ ì— ì •í†µí•œ 'AI ë²•ë¥  ë¹„ì„œ'ì…ë‹ˆë‹¤.
    ì‚¬ìš©ì(ì„ì°¨ì¸)ê°€ ì´í•´í•˜ê¸° ì‰½ë„ë¡ **í˜„ëŒ€ì ì¸ í•œêµ­ì–´(ì¡´ëŒ“ë§)**ë¡œ ë‹µë³€í•˜ì„¸ìš”.
    
    [ì—„ê²©í•œ ì œì•½ì‚¬í•­]
    1. **í•œì(Hanja), ì‚¬ìš© ê¸ˆì§€**: 'æ°‘æ³•', 'å‡ºé£' ê°™ì€ í•œìë¥¼ ì ˆëŒ€ ì“°ì§€ ë§ˆì„¸ìš”. í•œê¸€ë¡œë§Œ ì‘ì„±í•˜ì„¸ìš”.
    2. **ì‰¬ìš´ í‘œí˜„**: ë²•ë¥  ìš©ì–´ëŠ” ê´„í˜¸ ì•ˆì— ì„¤ëª…ì„ ë§ë¶™ì´ê±°ë‚˜, í’€ì–´ì„œ ì„¤ëª…í•˜ì„¸ìš”.
    
    [ì§€ì‹œì‚¬í•­]
    1. **ê·¼ê±° ê¸°ë°˜**: [ì°¸ê³  ë²•ë ¹]ì˜ ë‚´ìš©ì„ í† ëŒ€ë¡œ ë‹µë³€í•˜ì„¸ìš”. ì—†ëŠ” ì‚¬ì‹¤ì„ ì§€ì–´ë‚´ì§€ ë§ˆì„¸ìš”.
    2. **ê²°ë¡  ìš°ì„ **: ê°€ëŠ¥/ë¶ˆê°€ëŠ¥ ì—¬ë¶€ë¥¼ ë‘ê´„ì‹ìœ¼ë¡œ ë¨¼ì € ë§í•˜ì„¸ìš”.
    3. **ìœ„í—˜ ê²½ê³ **: ì„ì°¨ì¸ì—ê²Œ ë¶ˆë¦¬í•œ ë‚´ìš©ì´ë‚˜ ê°•í–‰ê·œì • ìœ„ë°˜(ë²•ì  ë¬´íš¨) ì‚¬í•­ì€ ê°•ë ¥íˆ ê²½ê³ í•˜ì„¸ìš”.
    4. **ë©´ì±…**: ë²•ì  íš¨ë ¥ì´ ì—†ëŠ” ì°¸ê³ ìš©ì„ì„ ë°íˆì„¸ìš”.

    [ì°¸ê³  ë²•ë ¹]
    {context}
    """
    
    prompt = ChatPromptTemplate.from_messages([
        ("system", system_prompt),
        ("human", "ì‚¬ìš©ì ì…ë ¥: {question}"),
    ])

    # ëª¨ë¸ ì„¤ì • (Solar ë˜ëŠ” Exaone)
    # í•œì ë¬¸ì œëŠ” ëª¨ë¸ íŠ¹ì„±ì¼ ìˆ˜ ìˆìœ¼ë‹ˆ, ë§Œì•½ ê³„ì† ê·¸ëŸ¬ë©´ exaone3.5:7.8b ì¶”ì²œ
    llm = ChatOllama(
        model="solar:10.7b", 
        temperature=0.1, 
        num_ctx=4096
    )

    chain = prompt | llm | StrOutputParser()
    
    try:
        # A. ìŠ¤íŠ¸ë¦¬ë° ëª¨ë“œ (í•œ ê¸€ìì”© ì¶œë ¥)
        if stream_output:
            full_response = ""
            for chunk in chain.stream({"context": context_text, "question": normalized_query}):
                print(chunk, end="", flush=True)
                full_response += chunk
            print() # ì¤„ë°”ê¿ˆ
            return full_response
            
        # B. ì¼ë°˜ ëª¨ë“œ (ì™„ì„± í›„ ë°˜í™˜ - Django ë“± ì›¹ ì—°ë™ ì‹œ ìœ ë¦¬)
        else:
            response = chain.invoke({"context": context_text, "question": normalized_query})
            return response
            
    except Exception as e:
        error_msg = f"âŒ ì˜¤ë¥˜ ë°œìƒ: {str(e)}"
        print(error_msg)
        return error_msg
```

```python
# =========================================================
# [Main Execution Block]
# =========================================================

# 1. Pinecone DB ì—°ê²° (ìµœì´ˆ 1íšŒ ì„¤ì •)
embedding_model = UpstageEmbeddings(model="solar-embedding-1-large-passage")
vector_store = PineconeVectorStore(
    pinecone_api_key=PINECONE_API_KEY,
    index_name="law-index",
    embedding=embedding_model
)
```

```python
# ---------------------------------------------------------
# â–¼â–¼â–¼ í…ŒìŠ¤íŠ¸í•  ì§ˆë¬¸ì„ ì—¬ê¸°ì— ì…ë ¥í•˜ì„¸ìš” â–¼â–¼â–¼
# user_input = "ì§‘ì£¼ì¸ì´ ì›”ì„¸ë¥¼ ì˜¬ë ¤ ë‹¬ë˜ìš”. ì œê°€ ì–´ë µë‹¤ê³  í•˜ë‹ˆê¹Œ ì €ë³´ê³  ë‹¤ìŒ ë‹¬ê¹Œì§€ ë‚˜ê°€ë¼ê³  í•˜ë„¤ìš”. ê³„ì•½ì„œì—ëŠ” ê·¸ëŸ° ì–˜ê¸°ê°€ ì—†ëŠ”ë° ì´ë ‡ê²Œ ì«“ê²¨ë‚˜ë„ ë˜ë‚˜ìš”?"
user_input = "ê³„ì•½ì„œì˜ ì´ ì¡°í•­ ê´œì°®ì„ê¹Œ? 'íŠ¹ì•½ì‚¬í•­: ì„ì°¨ì¸ì€ í‡´ì‹¤ ì‹œ ì²­ì†Œë¹„ 30ë§Œì›ì„ ë¬´ì¡°ê±´ ê³µì œí•œë‹¤.'" 
# ---------------------------------------------------------

print(f"ğŸ“¢ [Step 1] ì‚¬ìš©ì ì§ˆë¬¸: {user_input}")

# 2. ì •ê·œí™” (Normalization)
norm_query = normalize_query(user_input)
print(f"ğŸ”„ [Step 2] ê²€ìƒ‰ìš© ì¿¼ë¦¬ ë³€í™˜: {norm_query}")

# 3. ê²€ìƒ‰ ë° Reranking
retrieved_docs = legal_hybrid_retrieval(
    query=norm_query, 
    pinecone_store=vector_store,
    k_fetch=30,  # 1ì°¨ í›„ë³´
    k_final=5,   # ìµœì¢… ì„ ë³„
    score_threshold=0.25
)

if retrieved_docs:
    # 4. í¬ë§·íŒ…
    formatted_data = format_legal_docs(retrieved_docs)
    
    # (ì„ íƒ) ê²€ìƒ‰ëœ ì¡°í•­ ë¯¸ë¦¬ë³´ê¸°
    print(f"\nğŸ“š [Step 3] ìµœì¢… ì„ ë³„ëœ ë¬¸ì„œ {len(retrieved_docs)}ê°œ:")
    print(formatted_data['llm_context'])
    print("-" * 50)
    
    # 5. LLM ë‹µë³€ ìƒì„±
    final_answer = generate_legal_advice(
        normalized_query=norm_query, 
        context_text=formatted_data['llm_context'],
        stream_output=False # ë„ê¸°
    )
    
    print("=" * 50)
    print("ğŸ¤– [Step 4] AI ìµœì¢… ë‹µë³€")
    print("=" * 50)
    print(final_answer)

    # (ì›¹ ê°œë°œìš©) HTML ì½”ë“œ í™•ì¸
    # print("\n[Web Display HTML]")
    # print(formatted_data['display_html'])

else:
    print("âš ï¸ ê´€ë ¨ ë²•ë ¹ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. (ì ìˆ˜ ë¯¸ë‹¬)")
```
